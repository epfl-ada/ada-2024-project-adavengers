```python
import pathlib
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt

data_dir_path = pathlib.Path("../../data")
beer_mask = np.array(Image.open(data_dir_path/"beer_mask.png"))

text = """
beer ale lager pilsner IPA stout porter hops brewery fermentation draft craft malt yeast
pint amber blonde frothy golden refreshing bitter smooth dark crisp barley wheat seasonal
brewpub tasting cask fermentation suds bubbles hoppy barrel conditioning taproom mug foam
"""

wordcloud: WordCloud = WordCloud(
    background_color="white",
    mask=beer_mask,
    # contour_color="white",
    contour_width=20,
    colormap="autumn",  # beer-like colors
).generate(text)

plt.figure(figsize=(8, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Beer Word Cloud", fontsize=20, color="brown")
plt.show()
```
